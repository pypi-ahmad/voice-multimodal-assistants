# 04 - Market Insights and Report Generation

This notebook computes final metrics, builds chart artifacts, and renders a markdown report.

## Load Artifacts

```python
from __future__ import annotations

import json
from pathlib import Path

import polars as pl

from job_market_copilot.analysis.metrics import compute_metrics, save_charts, save_metrics
from job_market_copilot.config import get_settings
from job_market_copilot.reporting.report import render_report

settings = get_settings()
all_df = pl.read_parquet(settings.resolved_artifacts_dir / 'jobs_scored.parquet')
ai_df = pl.read_parquet(settings.resolved_artifacts_dir / 'ai_jobs_enriched.parquet')
all_df.shape, ai_df.shape
```

## Compute Metrics and Save Outputs

```python
metrics = compute_metrics(all_df, ai_df)
save_metrics(metrics, settings.resolved_artifacts_dir / 'metrics.json')
save_charts(ai_df, settings.resolved_artifacts_dir / 'charts')
report_path = settings.resolved_artifacts_dir / 'reports' / 'job_market_report.md'
render_report(metrics, settings.ollama_model, report_path)
metrics
```
