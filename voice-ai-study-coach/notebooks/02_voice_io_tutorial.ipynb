# 02 - Voice I/O Tutorial
Generate demo student audio, attach sidecar transcript, and inspect transcription.

```python
from pathlib import Path
from voice_study_coach.config import get_settings
from voice_study_coach.audio.synthesizer import ProceduralVoiceSynthesizer
from voice_study_coach.audio.transcriber import SidecarTranscriber
settings = get_settings()
synth = ProceduralVoiceSynthesizer(settings.audio_sample_rate, settings.audio_amplitude, settings.word_seconds, settings.pause_seconds)
transcriber = SidecarTranscriber()
```

```python
audio_path = settings.resolved_demo_audio_dir / 'notebook_student_question.wav'
text = 'Explain the rollback threshold for API incidents.'
seconds = synth.synthesize(text, audio_path)
audio_path.with_suffix('.txt').write_text(text, encoding='utf-8')
{'audio_path': audio_path.as_posix(), 'seconds': round(seconds, 3)}
```

```python
transcript = transcriber.transcribe(audio_path)
transcript
```
